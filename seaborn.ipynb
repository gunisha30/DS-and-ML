```python
from matplotlib import pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('Pokemon.csv', index_col=0, encoding='unicode_escape')
```

```python
plot1=plt.figure(1)
p=sns.lmplot(x='Attack',y='Defense',data=df)
plot1.plot(p)

plot2=plt.figure(2)
q=plt.scatter(df['Attack'],df['Defense'])
plot2.plot(q)

plt.show()
```
