```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv("Final 2.csv")

df=df.where(df['title']=='Data Scientist and Analyst')
df.dropna(inplace=True)

x=df['loc'].value_counts()
#print(df)
l=list(range(0,40))
df.set_index(pd.Index(l), inplace=True)

for i in range(len(x)):
    if (x[i]<5)==True:
        df.at[i,'loc']='Other'
        
#print(df)
labels = list(df['loc'].unique())
sizes = np.array(df['loc'].value_counts())

plt.pie(sizes, labels=labels)
plt.axis('equal')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy

labels = 'Frogs', 'Hogs', 'Dogs'
sizes = numpy.array([5860, 677, 3200])
colors = ['yellowgreen', 'gold', 'lightskyblue']

def absolute_value(val):
    a  = numpy.round(val/100.*sizes.sum(), 0)
    return a

plt.pie(sizes, labels=labels, colors=colors,
        autopct=absolute_value, shadow=True)

plt.axis('equal')
plt.show()
```
